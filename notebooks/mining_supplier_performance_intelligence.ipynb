# Mining Supplier Performance & Procurement Efficiency Intelligence

This notebook reviews supplier reliability, procurement efficiency, stockout exposure, incident impact, and site-level operational pressure from the frozen synthetic dataset.


```python
from pathlib import Path
import pandas as pd
ROOT = Path('..').resolve()
DATA = ROOT / 'data' / 'processed'
OUTPUTS = ROOT / 'outputs'
suppliers = pd.read_csv(DATA / 'suppliers.csv')
purchase_orders = pd.read_csv(DATA / 'purchase_orders.csv', parse_dates=['order_date','expected_delivery_date','actual_delivery_date'])
deliveries = pd.read_csv(DATA / 'deliveries.csv', parse_dates=['delivery_date'])
inventory_status = pd.read_csv(DATA / 'inventory_status.csv', parse_dates=['week'])
incidents = pd.read_csv(DATA / 'incidents.csv', parse_dates=['incident_date'])
site_summary = pd.read_csv(DATA / 'site_summary.csv')
supplier_summary = pd.read_csv(OUTPUTS / 'supplier_performance_summary.csv')
procurement_summary = pd.read_csv(OUTPUTS / 'procurement_cycle_summary.csv')
stockout_summary = pd.read_csv(OUTPUTS / 'stockout_risk_summary.csv')
incident_summary = pd.read_csv(OUTPUTS / 'incident_impact_summary.csv')
site_risk_summary = pd.read_csv(OUTPUTS / 'site_risk_summary.csv')
```
